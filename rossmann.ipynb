```python
import seaborn as sns
import matplotlib.pyplot as plt

train_encoded = pd.get_dummies(train, columns=['StateHoliday'])

# Correlation matrix for train data
correlation_matrix = train_encoded.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Train Data')
plt.show()
```

```python
store_encoded = pd.get_dummies(store, columns=['StoreType', 'Assortment', 'PromoInterval'])

# Correlation matrix for train data
store_correlation_matrix = store_encoded.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(store_correlation_matrix, annot = True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Store Data')
plt.show()
```

```python
merged_df = pd.merge(store, train, on='Store')

merged_encoded = pd.get_dummies(merged_df, columns=['StateHoliday', 'StoreType', 'Assortment', 'PromoInterval'])

# Correlation matrix for train data
merged_correlation_matrix = merged_encoded.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(merged_correlation_matrix, annot = True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Merged Data')
plt.show()
```

#### 1.	Sales (Umsatz): 

- **Customers (Kunden):** Sehr hohe positive Korrelation (~0.85). Dies bedeutet, dass mehr Kunden direkt mit höheren Umsätzen verbunden sind.
- **Promo (Promotion):** Moderate positive Korrelation (~0.30). Geschäfte, die Promotionen durchführen, tendieren dazu, höhere Umsätze zu erzielen.
- **Open (Geöffnet):** Positive Korrelation (~0.55). Umsätze werden nur erzielt, wenn das Geschäft geöffnet ist.
- **SchoolHoliday (Schulferien):** Schwache positive Korrelation (~0.10). Schulferien haben einen geringen positiven Einfluss auf den Umsatz.

StateHoliday (Staatliche Feiertage):

# Löschen der Einträge Store open ohne Verläufe/Kunden

```python
import pandas as pd
# Daten laden
train = pd.read_csv('rossmann-store-sales/train.csv')
store = pd.read_csv('rossmann-store-sales/store.csv')

# Daten zusammenführen
dataTrain = pd.merge(train, store, on='Store')

# Standardize StateHoliday column
dataTrain['StateHoliday'] = dataTrain['StateHoliday'].astype(str)

# Remove entries with zero sales and store open
dataTrain = dataTrain[~((dataTrain['Sales'] == 0) & (dataTrain['Open'] == 1))]

# Remove entries with zero customers and store open
dataTrain = dataTrain[~((dataTrain['Customers'] == 0) & (dataTrain['Open'] == 1))]

# Verify the entries are removed
print("\nTrain Data after removing entries with zero sales and store open:")
print(dataTrain[(dataTrain['Sales'] == 0) & (dataTrain['Open'] == 1)])

print("\nTrain Data after removing entries with zero customers and store open:")
print(dataTrain[(dataTrain['Customers'] == 0) & (dataTrain['Open'] == 1)])
```

# Heatmap mit fehlenden Werten

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fehlende Werte visualisieren
sns.heatmap(dataTrain.isnull(), cbar=False, cmap='viridis')
plt.show()
```

# Korrelationen

```python
numeric_data = dataTrain.select_dtypes(include=[np.number])

# Korrelationen und Heatmap
corr_matrix  = numeric_data.corr()

# Umwandeln der Matrix in ein langes Format
corr_pairs = corr_matrix.unstack()

# Entfernen der doppelten Werte (obere Dreieck der Matrix)
corr_pairs = corr_pairs.reset_index()
corr_pairs.columns = ['Variable1', 'Variable2', 'Correlation']
corr_pairs = corr_pairs[corr_pairs['Variable1'] != corr_pairs['Variable2']]

# Sortieren nach dem Betrag der Korrelation
corr_pairs['AbsCorrelation'] = np.abs(corr_pairs['Correlation'])
sorted_corr_pairs = corr_pairs.sort_values(by='AbsCorrelation', ascending=False)

# Filtern der bedeutenden Korrelationen (z.B. Betrag größer als 0.5)
significant_corr_pairs = sorted_corr_pairs[sorted_corr_pairs['AbsCorrelation'] > 0.5]

# Ausgabe der bedeutenden Korrelationen
print(significant_corr_pairs[['Variable1', 'Variable2', 'Correlation']])
```
